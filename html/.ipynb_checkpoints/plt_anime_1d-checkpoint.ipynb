```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import math,time

# å¾©ç¿’: ã‚¢ãƒ‹ãƒ¡ãªã—ã§ y = sin(x)ã‚’è¡¨ç¤º
def static_1d():
    x = np.linspace(0.0, 2.0 * math.pi, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.show()

# ã‚¢ãƒ‹ãƒ¡ã‚ã‚Š y = sin(k * x) ã‚’ k ã‚’è‰²ã€…å¤‰ãˆãªãŒã‚‰è¡¨ç¤º
def anime_1d():
    x = np.linspace(0.0, 2.0 * math.pi, 100)
    y = np.sin(x)
    # ä¸€åº¦ã ã‘ãƒ—ãƒ­ãƒƒãƒˆã—ã¦æ›²ç·šã‚’å¾—ã‚‹.
    # plotã®çµæžœ(ç·š)ã‚’å¤‰æ•°ã§å—ã‘å–ã‚‹ï¼Ž
    # å®Ÿã¯plotã¯ãƒªã‚¹ãƒˆã‚’è¿”ã™ã®ã§ï¼Œå·¦è¾ºã¯[line] = 
    # ã¨ã™ã‚‹
    [line] = plt.plot(x, y)
    for k in range(10):
        # ã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³ã™ã‚‹ã¨ãã¯ï¼Œplotã‚’å‘¼ã¶
        # plotã™ã‚‹ä»£ã‚ã‚Šã«ãƒ‡ãƒ¼ã‚¿ã‚’ã‚»ãƒƒãƒˆ(set_data)ã™ã‚‹
        line.set_data(x, np.sin(k * x))
        # å†æç”»ã—ã¦æ¬²ã—ã„ã‚¿ã‚¤ãƒŸãƒ³ã‚°ã§ï¼Œ yieldã‚’å‘¼ã¶
        # yield ã«ã¯ï¼Œå¤‰æ›´ã•ã‚ŒãŸæ›²ç·šã®ãƒªã‚¹ãƒˆã‚’æ¸¡ã™
        yield [line]

def anime_1dx():
    x = np.linspace(0.0, 2.0 * math.pi, 100)
    fig = plt.figure()
    plots = []
    for k in range(10):
        # ã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³ã™ã‚‹ã¨ãã¯ï¼Œplotã‚’å‘¼ã¶
        # plotã™ã‚‹ä»£ã‚ã‚Šã«ãƒ‡ãƒ¼ã‚¿ã‚’ã‚»ãƒƒãƒˆ(set_data)ã™ã‚‹
        plots.append(plt.plot(x, np.sin(k * x)))
    ani = animation.ArtistAnimation(fig, plots, interval=100)
    plt.show()
    return ani

# ãŠã¾ã˜ãªã„(ã‚³ãƒ”ãƒšã—ã¦ä¸‹ã•ã„ï¼Žã“ã‚Œã¾ã§ã®ç¯„å›²ã§ã¯
# ç†è§£ä¸èƒ½ã§ã™ã®ã§æ‚©ã‚€å¿…è¦ãªã—)
def go_animation(iterator, **kwargs):
    def anime_fun(*args):
        try:
            return next(iterator)
        except StopIteration:
            return []
    ani = animation.FuncAnimation(plt.gcf(), anime_fun, **kwargs)
    plt.show()

# sin(kx) ã®ã‚°ãƒ©ãƒ•ãŒå¾ã€…ã«å¤‰ã‚ã£ã¦ã„ã
#go_animation(anime_1d(), interval=100)
anime_1dx()

```
